```python
import keras
import os, cv2
import numpy as np
import pandas as pd
from keras import Input
import tensorflow as tf
from keras.models import Model
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
from keras.utils import to_categorical
import tensorflow.keras.backend as K
from sklearn.model_selection import train_test_split
from tensorflow.keras.layers import Dense, Activation, Flatten
from keras.layers import  BatchNormalization, Concatenate, Add, DepthwiseConv2D, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import StratifiedKFold
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import confusion_matrix
import seaborn as sns

# Load the images
data_path = '/content/drive/MyDrive/SZ outcomes 1'  # Path for generated images
data_dir_list = ['0', '1']
img_data_list = []
labels = []

for dataset in data_dir_list:
    img_list = os.listdir(os.path.join(data_path, dataset))  # Get list of images for the current dataset
    img_list = [img_name for img_name in img_list if img_name.endswith('.png')]
    print('Loaded the images of dataset-' + '{}\n'.format(dataset))
    for img_name in img_list:
        img = cv2.imread(os.path.join(data_path, dataset, img_name))
        img = cv2.resize(img, (128, 128))
        labels.append(dataset)
        img_data_list.append(img)


labels = [int(label) for label in labels]


label = np.array(labels)
img_data = np.array(img_data_list)
img_data = img_data.astype('float32')
img_data = img_data / 255
print(img_data.shape)

plt.imshow(img_data[30], cmap='gray')
plt.show()

# Data augmentation
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
num_classes = 2 # Number of classes in your dataset
x,y = shuffle(img_data, label, random_state=42)

from keras.layers import LeakyReLU
from keras.layers import *
def PSFAN(input, num_classes):
    conv1 = Conv2D(32, (3,3), dilation_rate=1, activation=LeakyReLU(alpha=0.02))(input)
    x = Conv2D(32, (1,1), padding='same')(conv1)
    x = Conv2D(32, (3,3), padding='same')(x)
    x = Conv2D(32, (1,1), padding='same')(x)
    a = Activation('sigmoid')(x)
    attention = Multiply()([a, conv1])
    add = Add()([x, attention])
    M = MaxPooling2D(pool_size=(3,3), padding='valid')(add)
    F = Flatten()(M)

    C = Concatenate()([F])

    D2 = Dense(128, activation=LeakyReLU(alpha=0.02))(C)
    D2 = Dense(64, activation=LeakyReLU(alpha=0.02))(D2)

    OUT = Dense(num_classes, activation='softmax')(D2)
    model = Model(inputs=[input], outputs=OUT, name="BaseModel")
    return model

input = Input(shape=(128, 128, 3), name='input')
model = PSFAN(input, num_classes)

min_samples = 1
class_counts = np.bincount(y)
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
# Lists to store training/validation loss and accuracy
train_loss = []
train_accuracy = []
val_loss = []
val_accuracy = []

# Define the lists to store accuracy scores
accuracy_scores = []
val_accuracy_list = []

for train_index, test_index in kfold.split(x, y):
    X_train, X_test = x[train_index], x[test_index]
    y_train, y_test = y[train_index], y[test_index]

    y_train = to_categorical(y_train, num_classes)
    y_test = to_categorical(y_test, num_classes)

    # Fit the model with data augmentation
    model.compile(optimizer=Adam(learning_rate=0.0001, weight_decay=0.0006), loss='binary_crossentropy', metrics=['accuracy'])
    history = model.fit(datagen.flow(X_train, y_train, batch_size=16),
                        steps_per_epoch=len(X_train) // 16,
                        epochs=50,
                        validation_data=(X_test, y_test),
                        verbose=1)

    # Append training/validation loss and accuracy
    train_loss.extend(history.history['loss'])
    train_accuracy.extend(history.history['accuracy'])
    val_loss.extend(history.history['val_loss'])
    val_accuracy_list.extend(history.history['val_accuracy'])

    val_acc = model.evaluate(X_test, y_test, verbose=0)[1]  # Get validation accuracy directly from model.evaluate
    print(f"Validation Accuracy: {val_acc:.2f}")
    val_accuracy_scores.append(val_acc)

    pred = model.predict(X_test)
    y_pred = np.argmax(pred, axis=1)
    y_true = np.argmax(y_test, axis=1)

    correct_count = sum(1 for pred, truth in zip(y_pred, y_true) if pred == truth)
    total_count = len(y_pred)

    accuracy = correct_count / total_count * 100
    print(f"Accuracy: {accuracy:.2f}")
    accuracy_scores.append(accuracy)

    # Plot training/validation loss
plt.figure(figsize=(12, 6))
plt.plot(train_loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Plot training/validation accuracy
plt.figure(figsize=(12, 6))
plt.plot(train_accuracy, label='Training Accuracy')
plt.plot(val_accuracy, label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

mean_accuracy = sum(accuracy_scores) / len(accuracy_scores)
mean_val_accuracy = sum(val_accuracy_scores) / len(val_accuracy_scores)

print(f"Mean Accuracy: {mean_accuracy:.2f}")
print(f"Mean Validation Accuracy: {mean_val_accuracy:.2f}")

# Calculate and plot confusion matrix
conf_matrix = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()

# Visualize predictions on a few sample images
num_samples = 5  # Number of samples to visualize
sample_indices = np.random.choice(len(X_test), num_samples, replace=False)

for idx in sample_indices:
    sample_image = X_test[idx]
    true_label = np.argmax(y_test[idx])
    pred_probs = model.predict(np.expand_dims(sample_image, axis=0))[0]
    predicted_label = np.argmax(pred_probs)
    predicted_class = "SZ Affected" if predicted_label == 1 else "Healthy"

    plt.imshow(sample_image)
    plt.title(f"True Label: {true_label}, Predicted Class: {predicted_class}, Probability: {pred_probs[predicted_label]:.4f}")
    plt.axis('off')
    plt.show()
```
